```python
#Droping the NA values from the Timestamp column
mental_health_cleaned = mental_health.dropna(subset=['Timestamp'])
```

```python
#Removing the I do not consent surveys
mental_cleaned_v1 = mental_health_cleaned[mental_health_cleaned['Please select one of the following options to indicate your consent to participate in this survey:'] != 'I do not consent to participate in this survey.']
```

```python
#Checking for null values after cleaning the data
mental_cleaned_v1.isnull().sum()
```

```python
#Converting date field to proper date format (yyyy/mm/dd)

mental_cleaned_v1['Timestamp'] = pd.to_datetime(mental_cleaned_v1['Timestamp']).dt.strftime('%Y/%m/%d')
```

```python
# Removing repetitive words throughout categories

mental_cleaned_v1['What is your age?'] = mental_cleaned_v1['What is your age?'].astype(str).str.replace('years', '')
mental_cleaned_v1['How many hours of sleep do you typically get per night?'] = mental_cleaned_v1['How many hours of sleep do you typically get per night?'].astype(str).str.replace('hours', '')
```

# **Exploratory Data Analysis**

## **Univariate Analysis**

```python
# Extracting the 'Physical_Activity_Frequency' column
physical_activity_column = mental_cleaned_v1['Physical_Activity_Frequency']

# Counting the frequency of each category
frequency_counts = physical_activity_column.value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(frequency_counts, labels=frequency_counts.index, autopct='%1.1f%%', colors=sns.color_palette('plasma'), startangle=90)
plt.title('Distribution of Physical Activity Frequency')

# Show the plot
plt.show()
```

## **Bivariate Analysis**

```python
columns_of_interest = ['Problem_Interest', 'Problem_Depression', 'Problem_Sleep', 'Problem_Energy', 'Problem_Appetite']

# Creating a DataFrame with selected columns
selected_mental_cleaned_v1 = mental_cleaned_v1[columns_of_interest]

# Creating a correlation matrix for categorical variables
corr_matrix = selected_mental_cleaned_v1.apply(lambda x: pd.factorize(x)[0]).corr(method='pearson')

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='magma', fmt='.2f', linewidths=.5)
plt.title('Heatmap of Categorical Variable Correlation')
plt.show()
```

```python
# Extracting the relevant columns
columns_of_interest = ['Country', 'Problem_Sleep']

# Creating a DataFrame with selected columns
selected_mental_cleaned_v1 = mental_cleaned_v1[columns_of_interest]

# Creating a crosstab for the two categorical variables
cross_tab = pd.crosstab(selected_mental_cleaned_v1['Country'], selected_mental_cleaned_v1['Problem_Sleep'])

# Plotting the clustered bar plot without stacking with a larger figsize
plt.figure(figsize=(16, 10))  # Adjust the size as needed
cross_tab.plot(kind='bar', colormap='viridis')
plt.title('Clustered Bar Plot of Country vs Problem Sleep')
plt.xlabel('Country')


# Show the plot
plt.tight_layout()
plt.show()
```

## **Multivariate Analysis**

```python
# Columns to consider for the chart
columns_for_chart_1 = [
    'Problem_Interest',
    'Problem_Depression',
    'Problem_Sleep',
    'Problem_Energy',
    'Problem_Appetite',
    'Problem_Self_Doubt',
    'Problem_Concentration',
    'Problem_Restlessness',
    'Problem_Suicidal_Thoughts'
]
```

```python
# Grouping by categories and counting the occurrences of each response
grouped_data = mental_cleaned_v1[columns_for_chart_1].apply(pd.Series.value_counts)

# Changing the order of bars
order_of_bars = ['Not at all', 'Rarely', 'Sometimes', 'Often', 'Almost always']
grouped_data = grouped_data.reindex(order_of_bars)
grouped_data
```

```python
# Transposing the grouped data for plotting
transposed_data = grouped_data.transpose()
transposed_data
```

```python
# Plotting the multi-variate clustered bar graph

ax = transposed_data.plot(kind='bar', stacked=False, figsize=(10, 6))

plt.xlabel('Categories')
plt.title('Clustered Column Chart')
plt.xticks(rotation=45)
plt.legend(labels=[f'{col}: {val}' for col, val in transposed_data.max().items()], bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
